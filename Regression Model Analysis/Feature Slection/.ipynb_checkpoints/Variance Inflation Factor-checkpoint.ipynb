```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats import diagnostic as diag
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.feature_selection import chi2
```

```python
dataset = pd.read_csv(r'/home/akhtar/Desktop/Ml Algorithms/Dataset/crops.csv')
dataset.head()
```

```python
corr = dataset.corr()

# display the correlation matrix
display(corr)

# plot the correlation heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
```

## " Dataset " Before extraction any features

```python
dataset_before = dataset

# the VFI does expect a constant term in the data, so we need to add one using the add_constant method
X1 = sm.tools.add_constant(dataset_before)

# create the series for both
series_before = pd.Series([variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])], index=X1.columns)

# display the series
print('DATA BEFORE')
print('-'*50)
display(series_before)
```

### " Dataset " After extraction of features ( 'Max_Temp', 'X_Cor', 'Y_Cor', 'Latitude', 'Longitude' )

```python
dataset_after = dataset.drop(['Max_Temp', 'X_COR', 'Y_COR', 'LATITUDE','LONGITUDE'], axis = 1)

X2 = sm.tools.add_constant(dataset_after)
series_after = pd.Series([variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])], index=X2.columns)

print('DATA AFTER')
print('-'*50)
display(series_after)
```

### " Dataset " After second extraction of features ( 'Min_Temp', 'Cloud_Coverage' )

```python
dataset_after2 = dataset_after.drop(['Min_Temp', 'Cloud_Coverage'], axis = 1)
X2 = sm.tools.add_constant(dataset_after2)
series_after = pd.Series([variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])], index=X2.columns)

print('DATA AFTER')
print('-'*50)
display(series_after)
```
