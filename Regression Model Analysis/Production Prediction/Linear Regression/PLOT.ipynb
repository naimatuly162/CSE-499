```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest
import statsmodels.api as sm
from statsmodels.stats import diagnostic as diag
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.feature_selection import chi2
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
dataset = pd.read_csv(r'/home/akhtar/Desktop/Ml Algorithms/Dataset/crops.csv')
dataset = dataset.drop(['Max_Temp', 'X_COR', 'Y_COR', 'LATITUDE','LONGITUDE', 'Min_Temp', 'Cloud_Coverage','Districts_N','Category_N', 'Crop_N', 'Year', 'Rainfall', 'Humidity', 'Wind_Speed', 'Bright_Sunshine', 'ALT'], axis = 1)

dataset.head()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y.values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Support Vector Regressor


```python
from sklearn.svm import SVR

regressor = SVR(kernel='rbf')
regressor.fit(X, y)
y_pred = regressor.predict(X)

# print("Accurecy: ", r2_score(y_test, y_pred))


plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=5)
regressor.fit(X, y)
y_pred = regressor.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

# Decision Tree

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=5)
regressor.fit(X, y)
y_pred = regressor.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

# KNN

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor()
regressor.fit(X, y)
y_pred = regressor.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

# ANN

```python
from sklearn.neural_network import MLPRegressor


regressor = MLPRegressor(hidden_layer_sizes = (100, 75, 50, 25), activation = 'relu', solver = 'sgd', learning_rate = 'adaptive')
regressor.fit(X, y)
y_pred = regressor.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```
